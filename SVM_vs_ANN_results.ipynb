```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
svm = pd.read_csv("/content/drive/MyDrive/Results/svm_results.csv")
ann = pd.read_csv("/content/drive/MyDrive/Results/ann_results.csv")
```

```python
ann = ann.rename(columns={"Dataset": "Image Condition"})
```

```python
comparison = svm.merge(
    ann,
    on="Image Condition",
    suffixes=("_SVM", "_ANN")
)

comparison
```

```python
x = np.arange(len(comparison))
width = 0.35

plt.figure(figsize=(10, 5))

plt.bar(x - width/2, comparison["Test Accuracy_SVM"], width, label="SVM")
plt.bar(x + width/2, comparison["Test Accuracy_ANN"], width, label="ANN")

plt.xticks(x, comparison["Image Condition"], rotation=30)
plt.ylabel("Accuracy")
plt.title("Test Accuracy Comparison")
plt.legend()

plt.tight_layout()
plt.show()
```
